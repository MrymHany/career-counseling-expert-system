```python
# Expert System for Career Counseling with Tkinter GUI (2025-2026)
# AI-CCE414 & ELE354 - Project 2
# Team Members: [Maryam Hany, Shaima saeed]

import tkinter as tk
from tkinter import ttk
from tkinter import scrolledtext
from tkinter import messagebox
import difflib

# --------------------- EXPANDED KNOWLEDGE BASE (Updated with BLS 2024-2034 and WEF 2025 Data) ---------------------
careers = {
    "Data Scientist": {
        "interests": ["technology", "analytics", "problem-solving", "data"],
        "skills": ["programming", "statistics", "machine-learning", "data-analysis", "analytical-thinking"],
        "academics": ["computer-science", "mathematics", "engineering", "statistics"],
        "trend_2025": "34% growth (BLS 2024-2034); High demand in AI/big data (WEF: 41% net growth).",
        "growth_rate": 34,
        "median_wage": "$112,590",
        "remote_friendly": True,
        "weights": {"interests": 0.3, "skills": 0.4, "academics": 0.3}
    },
    "Renewable Energy Engineer": {
        "interests": ["environment", "innovation", "engineering", "sustainability"],
        "skills": ["technical-design", "project-management", "sustainability", "environmental-stewardship"],
        "academics": ["engineering", "environmental-science"],
        "trend_2025": "38% net growth (WEF 2025); Ties to green transition.",
        "growth_rate": 38,
        "median_wage": "$51,860 (related)",
        "remote_friendly": False,
        "weights": {"interests": 0.35, "skills": 0.35, "academics": 0.3}
    },
    "Nurse Practitioner": {
        "interests": ["healthcare", "helping-others", "science"],
        "skills": ["patient-care", "diagnostics", "communication"],
        "academics": ["nursing", "biology", "medicine"],
        "trend_2025": "40% growth (BLS); Driven by aging population.",
        "growth_rate": 40,
        "median_wage": "$129,210",
        "remote_friendly": False,
        "weights": {"interests": 0.4, "skills": 0.3, "academics": 0.3}
    },
    "AI Specialist": {
        "interests": ["technology", "innovation", "ethics", "ai"],
        "skills": ["programming", "machine-learning", "data-analysis", "technological-literacy"],
        "academics": ["computer-science", "engineering"],
        "trend_2025": "82% net growth (WEF 2025); AI adoption surging.",
        "growth_rate": 82,
        "median_wage": "$140,910 (related)",
        "remote_friendly": True,
        "weights": {"interests": 0.3, "skills": 0.5, "academics": 0.2}
    },
    "Financial Advisor": {
        "interests": ["finance", "economics", "advising"],
        "skills": ["analytical", "communication", "risk-assessment"],
        "academics": ["business", "finance", "economics"],
        "trend_2025": "3% net growth (WEF); Steady in financial services.",
        "growth_rate": 3,
        "median_wage": "$90,400 (related)",
        "remote_friendly": True,
        "weights": {"interests": 0.3, "skills": 0.4, "academics": 0.3}
    },
    "Educator/Teacher": {
        "interests": ["education", "helping-others", "knowledge-sharing"],
        "skills": ["communication", "planning", "adaptability", "creative-thinking"],
        "academics": ["education", "liberal-arts"],
        "trend_2025": "17% growth for postsecondary (BLS); Tech integration rising.",
        "growth_rate": 17,
        "median_wage": "$79,940",
        "remote_friendly": True,
        "weights": {"interests": 0.4, "skills": 0.3, "academics": 0.3}
    },
    "Wind Turbine Service Technician": {
        "interests": ["environment", "technology", "engineering"],
        "skills": ["technical-design", "project-management", "sustainability"],
        "academics": ["engineering", "environmental-science"],
        "trend_2025": "50% growth (BLS); Boom in renewable energy.",
        "growth_rate": 50,
        "median_wage": "$62,580",
        "remote_friendly": False,
        "weights": {"interests": 0.35, "skills": 0.35, "academics": 0.3}
    },
    "Information Security Analyst": {
        "interests": ["technology", "security", "problem-solving"],
        "skills": ["networks-cybersecurity", "analytical-thinking", "technological-literacy"],
        "academics": ["computer-science", "engineering"],
        "trend_2025": "29% growth (BLS); Cybersecurity concerns rising.",
        "growth_rate": 29,
        "median_wage": "$124,910",
        "remote_friendly": True,
        "weights": {"interests": 0.3, "skills": 0.4, "academics": 0.3}
    },
    "Sustainability Specialist": {
        "interests": ["environment", "innovation", "ethics"],
        "skills": ["environmental-stewardship", "systems-thinking", "creative-thinking"],
        "academics": ["environmental-science", "business"],
        "trend_2025": "33% net growth (WEF); Green transition focus.",
        "growth_rate": 33,
        "median_wage": "N/A (emerging)",
        "remote_friendly": True,
        "weights": {"interests": 0.4, "skills": 0.3, "academics": 0.3}
    },
    "Big Data Specialist": {
        "interests": ["technology", "analytics", "data"],
        "skills": ["programming", "statistics", "ai-big-data", "analytical-thinking"],
        "academics": ["computer-science", "mathematics"],
        "trend_2025": "27% net growth (WEF); Big data trends exploding.",
        "growth_rate": 27,
        "median_wage": "$112,590 (related)",
        "remote_friendly": True,
        "weights": {"interests": 0.3, "skills": 0.4, "academics": 0.3}
    }
}

# Options lists
interests_options = ["technology", "environment", "healthcare", "finance", "education", "innovation", "problem-solving", "helping-others", "science", "economics", "ethics", "knowledge-sharing", "data", "ai", "security", "sustainability"]
skills_options = ["programming", "statistics", "machine-learning", "technical-design", "project-management", "sustainability", "patient-care", "diagnostics", "communication", "analytical", "risk-assessment", "planning", "adaptability", "data-analysis", "networks-cybersecurity", "analytical-thinking", "technological-literacy", "environmental-stewardship", "systems-thinking", "creative-thinking", "ai-big-data"]
academics_options = ["computer-science", "mathematics", "engineering", "environmental-science", "nursing", "biology", "medicine", "business", "finance", "economics", "education", "liberal-arts", "statistics"]
preferences_options = ["high_growth", "remote_work"]

# --------------------- INFERENCE ENGINE ---------------------
def fuzzy_match(user_list, target_list):
    matches = 0
    for u in user_list:
        for t in target_list:
            if difflib.SequenceMatcher(None, u.lower(), t.lower()).ratio() > 0.6:
                matches += 1
                break
    return matches

def calculate_scores(user_interests, user_skills, user_academics, preferences):
    results = {}
    total_scores = []
    for career, info in careers.items():
        i_match = fuzzy_match(user_interests, info["interests"])
        s_match = fuzzy_match(user_skills, info["skills"])
        a_match = fuzzy_match(user_academics, info["academics"])
        
        i_score = (i_match / len(info["interests"])) * info["weights"]["interests"] * 100 if info["interests"] else 0
        s_score = (s_match / len(info["skills"])) * info["weights"]["skills"] * 100 if info["skills"] else 0
        a_score = (a_match / len(info["academics"])) * info["weights"]["academics"] * 100 if info["academics"] else 0
        
        base_score = i_score + s_score + a_score
        
        trend_bonus = min(info["growth_rate"] * 0.2, 20) if info["growth_rate"] >= 20 else 0
        
        pref_bonus = 0
        if 'high_growth' in preferences and info["growth_rate"] > 30:
            pref_bonus += 15
        if 'remote_work' in preferences and info.get("remote_friendly", False):
            pref_bonus += 10
        
        total_score = base_score + trend_bonus + pref_bonus
        total_scores.append(total_score)
        results[career] = {"score": total_score, "info": info}
    avg_score = sum(total_scores) / len(total_scores) if total_scores else 0
    return sorted(results.items(), key=lambda x: x[1]["score"], reverse=True), avg_score

# --------------------- GUI DESIGN ---------------------
root = tk.Tk()
root.title("Career Counseling Expert System - Faculty of Engineering at Shoubra")
root.geometry("1400x850")
root.configure(bg="#F9FAFB")  # Light gray background

# Title
title_frame = tk.Frame(root, bg="#F9FAFB")
title_frame.pack(pady=30)
title_label = tk.Label(title_frame, text="Career Counseling Expert System", font=("Helvetica Neue", 28, "bold"), bg="#F9FAFB", fg="#DB2777")
title_label.pack()
subtitle_label = tk.Label(title_frame, text="Faculty of Engineering at Shoubra | 2025-2026 Trends", font=("Helvetica Neue", 14), bg="#F9FAFB", fg="#64748B")
subtitle_label.pack(pady=5)

# Style for tabs
style = ttk.Style()
style.theme_use('clam')
style.configure('TNotebook', background='#F9FAFB')
style.configure('TNotebook.Tab', background='#E2E8F0', foreground="#DB2777", padding=[20, 12], font=('Helvetica Neue', 13, 'bold'))
style.map('TNotebook.Tab', background=[('selected', '#DB2777')], foreground=[('selected', 'white')])

notebook = ttk.Notebook(root)
notebook.pack(pady=20, padx=40, fill=tk.BOTH, expand=True)

# Input Tab
input_tab = ttk.Frame(notebook)
notebook.add(input_tab, text="Input Parameters")

input_frame = tk.Frame(input_tab, bg="#F9FAFB")
input_frame.pack(pady=30, padx=40, fill=tk.BOTH, expand=True)

def create_pro_listbox(parent, label_text):
    tk.Label(parent, text=label_text, font=("Helvetica Neue", 13, "bold"), bg="#F9FAFB", fg="#DB2777").grid(row=0, column=parent.grid_size()[0], padx=30, pady=15, sticky=tk.W)
    frame = tk.Frame(parent, bg="#F9FAFB")
    frame.grid(row=1, column=parent.grid_size()[0]-1, padx=30, pady=10)
    scroll = tk.Scrollbar(frame, bg="#E2E8F0")
    scroll.pack(side=tk.RIGHT, fill=tk.Y)
    lb = tk.Listbox(frame, selectmode=tk.MULTIPLE, height=14, yscrollcommand=scroll.set, exportselection=0, bg="white", fg="#1F2937", selectbackground="#F472B6", font=("Helvetica Neue", 11), relief=tk.FLAT, highlightthickness=0)
    scroll.config(command=lb.yview)
    lb.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    return lb

interests_lb = create_pro_listbox(input_frame, "Select Interests")
for item in interests_options:
    interests_lb.insert(tk.END, item)

skills_lb = create_pro_listbox(input_frame, "Select Skills")
for item in skills_options:
    skills_lb.insert(tk.END, item)

academics_lb = create_pro_listbox(input_frame, "Select Academic Background")
for item in academics_options:
    academics_lb.insert(tk.END, item)

preferences_lb = create_pro_listbox(input_frame, "Select Preferences")
for item in preferences_options:
    preferences_lb.insert(tk.END, item)

# Buttons with hover effect
button_frame = tk.Frame(input_tab, bg="#F9FAFB")
button_frame.pack(pady=40)

def on_enter(e, btn):
    btn.config(background='#F472B6')

def on_leave(e, btn):
    btn.config(background='#DB2777')

btn_get = tk.Button(button_frame, text="Generate Recommendations", font=("Helvetica Neue", 16, "bold"), bg="#DB2777", fg="white", relief=tk.FLAT, padx=40, pady=15, cursor="hand2")
btn_get.pack(side=tk.LEFT, padx=40)
btn_get.bind("<Enter>", lambda e: on_enter(e, btn_get))
btn_get.bind("<Leave>", lambda e: on_leave(e, btn_get))

btn_clear = tk.Button(button_frame, text="Clear Selection", font=("Helvetica Neue", 16, "bold"), bg="#64748B", fg="white", relief=tk.FLAT, padx=40, pady=15, cursor="hand2")
btn_clear.pack(side=tk.LEFT, padx=40)
btn_clear.bind("<Enter>", lambda e: btn_clear.config(background='#475569'))
btn_clear.bind("<Leave>", lambda e: btn_clear.config(background='#64748B'))

# Results Tab
results_tab = ttk.Frame(notebook)
notebook.add(results_tab, text="Recommendations")

result_text = scrolledtext.ScrolledText(results_tab, font=("Helvetica Neue", 12), bg="white", fg="#1F2937", wrap=tk.WORD, relief=tk.FLAT, bd=0, highlightthickness=0, insertbackground="#DB2777")
result_text.pack(pady=40, padx=40, fill=tk.BOTH, expand=True)

# Status Bar
status_var = tk.StringVar(value="Ready | Faculty of Engineering at Shoubra")
status_bar = tk.Label(root, textvariable=status_var, bd=1, relief=tk.SUNKEN, anchor=tk.W, bg="#DB2777", fg="white", font=("Helvetica Neue", 10))
status_bar.pack(side=tk.BOTTOM, fill=tk.X)

# Functions
def clear_selections():
    interests_lb.selection_clear(0, tk.END)
    skills_lb.selection_clear(0, tk.END)
    academics_lb.selection_clear(0, tk.END)
    preferences_lb.selection_clear(0, tk.END)
    result_text.delete(1.0, tk.END)
    status_var.set("Cleared | Ready for new selection")

btn_clear.config(command=clear_selections)

def get_recommendations():
    user_interests = [interests_lb.get(idx) for idx in interests_lb.curselection()]
    user_skills = [skills_lb.get(idx) for idx in skills_lb.curselection()]
    user_academics = [academics_lb.get(idx) for idx in academics_lb.curselection()]
    preferences = [preferences_lb.get(idx) for idx in preferences_lb.curselection()]
    
    if not (user_interests or user_skills or user_academics):
        messagebox.showwarning("Input Required", "Please select at least one item from Interests, Skills, or Academic Background.")
        return
    
    status_var.set("Analyzing your profile...")
    root.update()
    
    top_careers, avg_score = calculate_scores(user_interests, user_skills, user_academics, preferences)
    
    result_text.delete(1.0, tk.END)
    result_text.insert(tk.END, "Career Recommendations (2025-2026 Trends)\n\n", "title")
    
    found = False
    for career, data in top_careers[:5]:
        if data["score"] >= 50:
            found = True
            result_text.insert(tk.END, f"{career}\n", "career")
            result_text.insert(tk.END, f"Match Score: {data['score']:.1f}%\n", "score")
            result_text.insert(tk.END, f"Market Outlook: {data['info']['trend_2025']}\n")
            result_text.insert(tk.END, f"Growth Rate: {data['info']['growth_rate']}%\n")
            result_text.insert(tk.END, f"Median Annual Wage: {data['info']['median_wage']}\n")
            result_text.insert(tk.END, f"Next Steps: Develop skills in {', '.join(data['info']['skills'][:3])}\n")
            result_text.insert(tk.END, "─" * 70 + "\n\n")
    
    if not found:
        result_text.insert(tk.END, "No strong matches found. Consider high-growth fields like AI or Renewables.\n")
    
    result_text.insert(tk.END, f"Average Match Score: {avg_score:.1f}%\n\n")
    result_text.insert(tk.END, "Data Source: U.S. Bureau of Labor Statistics (BLS) & World Economic Forum (WEF) 2025 Reports\n")
    result_text.insert(tk.END, "Disclaimer: This is an educational expert system. Consult a professional career advisor.\n")
    
    notebook.select(results_tab)
    status_var.set("Analysis Complete")

btn_get.config(command=get_recommendations)

# Styling tags for results
result_text.tag_config("title", font=("Helvetica Neue", 20, "bold"), foreground="#DB2777", justify="center")
result_text.tag_config("career", font=("Helvetica Neue", 16, "bold"), foreground="#DB2777")
result_text.tag_config("score", font=("Helvetica Neue", 14, "bold"), foreground="#F472B6")

root.mainloop()
```
